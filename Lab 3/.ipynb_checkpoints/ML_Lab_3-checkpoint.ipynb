## Постановка задачи

Воспользуемся датасетом ```sms_spam```, расположенном в директории [Data](https://drive.google.com/drive/folders/1YAMe7MiTxA-RSSd8Ex2p-L0Dspe6Gs4L), для решения задачи фильтрации спама.


Для выполнения поставленной задачи будем следовать следующему плану:
1. Приведение к нижнему регистру
2. Удаление стоп-слов
3. Токенизация
4. Лемматизация
5. Векторизация и настройка гиперпараметров
6. Построение модели
7. Оценка качества модели
8. Выявление лучших моделей

Используем следующие алгоритмы векторизации:
1. Мешок n-грамм
2. TF-IDF

В качестве классификатора используем классификацию с помошью метода опорных векторов. В качестве метрик для оценки качества модели используем accuracy, precision, recall, f1-score 

````python
import nltk
import pandas as pd
import sklearn
import numpy as np
import string
import matplotlib.pyplot as plt
import warnings
import seaborn as sns

from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from termcolor import colored
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, precision_recall_fscore_support, confusion_matrix
from sklearn.naive_bayes import MultinomialNB
from sklearn import svm
from gensim.models.word2vec import Word2Vec

warnings.filterwarnings('ignore')
nltk.download("stopwords", quiet=True)
nltk.download('punkt', quiet=True)
nltk.download("wordnet", quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
````

**0. Начальная обработка данных**

````python
df = pd.read_csv("sms_spam.csv", sep=",")
df
````

Заменим значения меток на 0 - 'ham', 1 - 'spam', а также уберем знаки препинания из текста

````python
df = df.rename(columns={'type': 'label'})
df.replace({"label": {"ham": 0, "spam": 1}}, inplace=True)
print(colored("Знаки препинания: ", "red"), string.punctuation)
df.text = df.text.str.replace(rf"[{string.punctuation}]", "", regex=True) 
df
````

**1. Приведение к нижнему регистру**

````python
df.text = df.text.str.lower()
df
````

**2. Удаление стоп-слов**

````python
stop_list = stopwords.words("english")
print(colored("Stop words: ", "red"))
print(*stop_list[:10], sep=", ", end=" ...")
df.text = df.text.apply(lambda s: " ".join([word for word in s.split() if word not in stop_list]))
df
````

**3. Токенизация**

````python
df.text = df.text.apply(word_tokenize)
df
````

**4. Лемматизация**

````python
wnl = WordNetLemmatizer()

def get_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

df.text = df.text.apply(lambda s: " ".join([wnl.lemmatize(word, get_pos(word)) for word in s]))
df
````

**5-7. Векторизация и настройка гиперпараметров, построение модели, оценка качества**

````python
df_prepared = df.copy(deep=True)
x_train, x_test, y_train, y_test = train_test_split(df_prepared.text,
                                                    df_prepared.label,
                                                    train_size=0.8)

modelNB = MultinomialNB() # Мультиномиальный наивный байесовский классификатор
modelSVC = svm.SVC() # Support vector classifier
````

````python
def create_x_ticks(df_index: pd.MultiIndex):
    return ["{}\nclass={}\nsupport={}".format(item[0].replace(' ', '\n'), item[1], item[2]) for item in df_index]
````

**1) Мешок слов**

````python
ngram_max = 5
ngram_ranges = [(i, j) for i in range(1, ngram_max + 1) for j in range(i, ngram_max + 1)]
df_result_ngram = pd.DataFrame()
for ngram_range in ngram_ranges:
    #Векторизация
    cv = CountVectorizer(ngram_range=ngram_range)
    x_train_ngram = cv.fit_transform(x_train)
    x_test_ngram = cv.transform(x_test)
    vectorizer_name = f"{cv.__class__.__name__} ngram=({ngram_range[0]},{ngram_range[1]})"

    #Обучение
    # modelNB.fit(x_train_ngram, y_train)
    modelSVC.fit(x_train_ngram, y_train)

    #Тест
    # y_pred_ngram = modelNB.predict(x_test_ngram)
    y_pred_ngram = modelSVC.predict(x_test_ngram)
    report_ngram = classification_report(y_test, y_pred_ngram, output_dict=True)
    report_ngram.update({"accuracy": {
        "precision": None,
        "recall": None,
        "f1-score": report_ngram["accuracy"],
        "support": report_ngram["macro avg"]["support"]}})

    #Результаты
    result_ngram = pd.DataFrame(report_ngram).T
    result_ngram.reset_index(level=0, inplace=True)
    result_ngram = result_ngram.rename(columns={"index": "label"})
    result_ngram["name"] = vectorizer_name
    result_ngram.set_index(["name", "label", "support"], inplace=True)
    df_result_ngram = pd.concat([df_result_ngram, result_ngram])
````

````python
df_result_ngram
````

Определим значение гиперпараметра векторайзера, при котором он дает наиболее правильные результаты. Сравнение будем производить по значению параметра ```macro_avg```

````python
df_result_ngram_plot = pd.DataFrame()
for column in df_result_ngram.columns:
    df_result_ngram_avg = df_result_ngram.loc[df_result_ngram.index.get_level_values("label") == "macro avg"]
    df_result_ngram_classes = df_result_ngram.loc[
        df_result_ngram.index.get_level_values("label").isin(["0", "1", "macro avg"])]
    index_max = df_result_ngram_avg[column].idxmax()
    df_result_ngram_plot = pd.concat([df_result_ngram_plot, df_result_ngram_classes.loc[[index_max[0]]]])

df_result_ngram_plot.drop_duplicates(inplace=True)
# print(df_result_ngram_plot.index)
df_result_ngram_plot = df_result_ngram_plot.loc[df_result_ngram_plot.index.get_level_values("label") == "macro avg"]
df_result_ngram_plot = df_result_ngram_plot.loc[df_result_ngram_plot.index.get_level_values("name") == "CountVectorizer ngram=(1,1)"]
df_result_ngram_plot
````

````python
# Визуализация CountVectorizer
ngram_plot = df_result_ngram_plot.plot(rot=0, figsize=(25, 12), ylim=(0.80, 1.02), kind='bar', fontsize=20)
ngram_plot.legend(fontsize=20)
ngram_plot.grid(alpha=0.5, ls='--')
ngram_set_xticks = ngram_plot.set_xticklabels(create_x_ticks(df_result_ngram_plot.index))
````

**2) TF-IDF**

````python
tfidf_max_size = 5
df_result_tfidf = pd.DataFrame()
tfidf_ranges = [(i, j) for i in range(1, tfidf_max_size) for j in range(i, tfidf_max_size)]

for tfidf_range in tfidf_ranges:
    #Векторизация
    tfidf_vectorizer = TfidfVectorizer(ngram_range=tfidf_range)
    x_train_tfidf = tfidf_vectorizer.fit_transform(x_train)
    
    
    x_test_tfidf = tfidf_vectorizer.transform(x_test)
    vectorizer_name = f"{tfidf_vectorizer.__class__.__name__} ngram=({tfidf_range[0]},{tfidf_range[1]}) "
    
    
    #Обучение
    # modelNB.fit(x_train_tfidf, y_train)
    modelSVC.fit(x_train_tfidf, y_train)
    
    #Тест
    # y_pred_tfidf = modelNB.predict(x_test_tfidf)
    y_pred_tfidf = modelSVC.predict(x_test_tfidf)
    report_tfidf = classification_report(y_test, y_pred_tfidf, output_dict=True)
    report_tfidf.update({"accuracy": {
        "precision": None,
        "recall": None,
        "f1-score": report_tfidf["accuracy"],
        "support": report_tfidf["macro avg"]["support"]}})
    
    #Результаты
    result_tfidf = pd.DataFrame(report_tfidf).T
    result_tfidf.reset_index(level=0, inplace=True)
    result_tfidf = result_tfidf.rename(columns={"index": "label"})
    result_tfidf["name"] = vectorizer_name
    result_tfidf.set_index(["name", "label", "support"], inplace=True)
    df_result_tfidf = pd.concat([df_result_tfidf, result_tfidf])
````

Определим значение гиперпараметра векторайзера, при котором он дает наиболее правильные результаты. Сравнение будем производить по значению параметра ```macro_avg```

````python
df_result_tfidf_plot = pd.DataFrame()
for column in df_result_tfidf.columns:
    df_result_tfidf_avg = df_result_tfidf.loc[
        df_result_tfidf.index.get_level_values("label") == "macro avg"]
    df_result_tfidf_classes = df_result_tfidf.loc[
        df_result_tfidf.index.get_level_values("label").isin(["0", "1", "macro avg"])]
    index_max = df_result_tfidf_avg[column].idxmax()
    df_result_tfidf_plot = pd.concat(
        [df_result_tfidf_plot, df_result_tfidf_classes.loc[[index_max[0]]]])
    
df_result_tfidf_plot.drop_duplicates(inplace=True)
df_result_tfidf_plot = df_result_tfidf_plot.loc[df_result_tfidf_plot.index.get_level_values("label") == "macro avg"]
df_result_tfidf_plot = df_result_tfidf_plot.loc[df_result_tfidf_plot.index.get_level_values("name") == "TfidfVectorizer ngram=(1,1) "]
df_result_tfidf_plot
````

````python
# Визуализация TF-IDF
tfidf_plot = df_result_tfidf_plot.plot(rot=0, figsize=(25, 12), ylim=(0.80, 1.02), kind='bar', fontsize=20)
tfidf_plot.legend(fontsize=20)
tfidf_plot.grid(alpha=0.5, ls='--')
tfidf_set_xticks = tfidf_plot.set_xticklabels(create_x_ticks(df_result_tfidf_plot.index))
````

**Примеры векторизации с помощью мешка слов и TF-IDF**

````python
corpus = [
     'This is the first document.',
     'This document is the second document.',
     'And this is the third one.',
     'Is this the first document?',
]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print(colored("Dictionary:", "red"), vectorizer.get_feature_names_out(), end="\n")
print(colored("The sparse matrix of each sentence in the corpus", "red"), X, sep="\n", end="\n")
print(colored("The usual matrix of the corpus", "red"), X.toarray(), sep="\n")
````

````python
corpus = [
     'This is the first document.',
     'This document is the second document.',
     'And this is the third one.',
     'Is this the first document?',
]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
print(colored("Dictionary:", "red"), vectorizer.get_feature_names_out(), end="\n")
print(colored("The sparse matrix of each sentence in the corpus", "red"), np.round(X, 2), sep="\n", end="\n")
print(colored("The usual matrix of the corpus", "red"), np.round(X.toarray(), 2), sep="\n")
````
